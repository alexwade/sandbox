# Query the Covid19 dataset on GBQ

## Get a list of doctypes and counts

```python
%%bigquery bytype

SELECT distinct(type), count(id) as count
FROM `covid-19-dimensions-ai.data.publications`
group by type order by count desc
```

```python
 bytype
```

```python
alt.Chart(bytype).mark_bar().encode(
    x='count',
    y=alt.Y('type', sort='-x')
)
```

## What are the classification model outputs for type=proceeding?

```python
%%bigquery doctypes

SELECT distinct(document_type.classification) doctype, count(id) as count
FROM `covid-19-dimensions-ai.data.publications`
where type = "proceeding"
group by doctype order by count desc
```

```python
doctypes
```

```python
alt.Chart(doctypes).mark_bar().encode(
    x='count',
    y=alt.Y('doctype', sort='-x')
)
```
